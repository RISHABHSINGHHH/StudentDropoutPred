```python
import pandas as pd
```

```python
dataset = pd.read_csv("student dropout.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum().sort_values(ascending=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
```

```python
categorical_cols = dataset.select_dtypes(include=['object', 'category']).columns
encoder=LabelEncoder()

dataset[categorical_cols]=encoder.fit_transform(categorical_cols)
```

```python
X = dataset.drop('Dropped_Out', axis=1)
Y = dataset['Dropped_Out']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Initialize the model
model = Sequential()
model.add(Dense(units=64, activation='tanh', input_shape=(X_train_scaled.shape[1],)))

# Add hidden layers
model.add(Dense(units=64, activation='tanh'))
model.add(Dense(units=64, activation='tanh'))

# Add output layer
model.add(Dense(units=1, activation='sigmoid'))
```

```python
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# Train the model
history = model.fit(X_train_scaled, Y_train, epochs=40, batch_size=32, validation_split=0.2)
```

```python
# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test_scaled, Y_test)


print(f'Accuracy: {test_accuracy}')
```

```python
# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Validation'], loc='upper left')
#plt.savefig('ANN-accuracy.pdf')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Validation'], loc='upper left')
#plt.savefig('ANN-loss.pdf')
plt.show()
```

```python
param_grid = {
    'n_neighbors': [3, 5, 7, 9],   # Different values of k
    'weights': ['uniform', 'distance'],  # Whether to use uniform weights or distance-based
    'metric': ['euclidean', 'manhattan', 'minkowski']  # Distance metric to use
}
```

```python
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy')
```

```python
grid_search.fit(X_train, Y_train)
```

```python
print(f"Best parameters found: {grid_search.best_params_}")
print(f"Best cross-validation accuracy: {grid_search.best_score_:.2f}")

# Step 9: Evaluate on the test set with the best parameters
best_knn = grid_search.best_estimator_
test_accuracy = best_knn.score(X_test, Y_test)
print(f"Test accuracy: {test_accuracy * 100:.2f}%")
```

```python
param_grid = {
    'n_neighbors': np.arange(1, 20),
    'weights': ['uniform', 'distance']
}

# Step 9: Set up the GridSearchCV and fit it (same as before)
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy', return_train_score=True)
grid_search.fit(X_train, Y_train)

# Step 10: Extract the results into a pandas DataFrame
results = pd.DataFrame(grid_search.cv_results_)

# Step 11: Pivot the DataFrame to get a matrix-like structure for heatmap
pivot_table = results.pivot(index='param_n_neighbors', columns='param_weights', values='mean_test_score')

# Step 12: Plot a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, cmap='viridis', cbar=True)
plt.title('Hyperparameter Tuning: Accuracy Heatmap')
plt.xlabel('Weights')
plt.ylabel('Number of Neighbors (k)')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
param_grid = {
    'n_neighbors': np.arange(1, 20),
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean','minkowski','manhattan']
}

# Step 6: Set up the GridSearchCV
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy', return_train_score=True)

# Step 7: Fit the GridSearchCV on the training data
grid_search.fit(X_train, Y_train)

# Step 8: Extract the results into a pandas DataFrame
results = pd.DataFrame(grid_search.cv_results_)

# Step 9: Convert categorical columns to numerical for plotting
results['weights'] = results['param_weights'].map({'uniform': 0, 'distance': 1})
results['metric'] = results['param_metric'].map({'euclidean': 0, 'manhattan': 1, 'minkowski': 2})

# Step 10: Set up 3D plotting
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Step 11: Extract hyperparameters and score for plotting
x = results['param_n_neighbors'].astype(int)
y = results['weights'].astype(int)  # 0 = uniform, 1 = distance
z = results['metric'].astype(int)  # 0 = euclidean, 1 = manhattan
accuracy = results['mean_test_score']

# Step 12: Plot the 3D scatter plot
img = ax.scatter(x, y, z, c=accuracy, cmap='viridis', s=100)

# Step 13: Customize the plot
ax.set_xlabel('Number of Neighbors (k)')
ax.set_ylabel('Weights (0=uniform, 1=distance)')
ax.set_zlabel('Metric (0=euclidean, 1=manhattan, 2=minkowski)')
ax.set_title('3D Plot of Hyperparameters and Cross-Validation Accuracy')

# Add color bar
cbar = fig.colorbar(img, ax=ax, pad=0.1)
cbar.set_label('Cross-Validation Accuracy')

#plt.savefig('knn_hyperparameter_tuning.pdf', dpi=300)

plt.show()
```
